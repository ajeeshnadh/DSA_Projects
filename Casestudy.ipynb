```python
# Example: unify country names across dataframes
def standardize_country_names(df, col):
    mapping = {
        "United States of America": "USA",
        "United Kingdom": "UK",
        "Russian Federation": "Russia",
        "Republic of Korea": "South Korea",
        "Viet Nam": "Vietnam",
        "United States": "USA" # Add mapping for "United States"
    }
    # Ensure the column exists before applying replace
    if col in df.columns:
      df[col] = df[col].replace(mapping)
    else:
      print(f"Warning: Column '{col}' not found in DataFrame.")
    return df

# Handle the structure of internet_users_df before standardization
# The actual country names are in the first row (index 0) of the 'country' column
internet_users_df.columns = internet_users_df.iloc[0] # Set the first row as the new header
internet_users_df = internet_users_df[1:].reset_index(drop=True) # Remove the first row and reset index
internet_users_df = internet_users_df.rename(columns={'country': 'Country'}) # Rename the first column to 'Country'


population_df = standardize_country_names(population_df, "country") # Use 'country' for population_df
gdp_df = standardize_country_names(gdp_df, "Country") # Use 'Country' for gdp_df
internet_users_df = standardize_country_names(internet_users_df, "Country") # Use 'Country' for internet_users_df after restructuring
literacy_df = standardize_country_names(literacy_df, "country") # Use 'country' for literacy_df

# Display head of standardized dataframes
print("Population DataFrame after standardization:")
display(population_df.head())
print("\nGDP DataFrame after standardization:")
display(gdp_df.head())
print("\nInternet Users DataFrame after standardization:")
display(internet_users_df.head())
print("\nLiteracy Rate DataFrame after standardization:")
display(literacy_df.head())
```

```python
# Ensure numeric columns are in correct format

# For population_df, the column is 'population'
population_df["population"] = pd.to_numeric(population_df["population"], errors="coerce")

# For gdp_df, the column is 'GDP'
gdp_df["GDP"] = pd.to_numeric(gdp_df["GDP"], errors="coerce")

# For internet_users_df, the data is spread across columns and rows.
# We need to melt the DataFrame and convert the 'internet_users' values to numeric.
internet_users_df_melted = internet_users_df.melt(id_vars=['Country'], var_name='metric', value_name='value')
internet_users_df_numeric = internet_users_df_melted[internet_users_df_melted['metric'] == 'internet_users'].rename(columns={'value': 'InternetUsers'})
internet_users_df_numeric = internet_users_df_numeric[['Country', 'InternetUsers']]
internet_users_df_numeric["InternetUsers"] = pd.to_numeric(internet_users_df_numeric["InternetUsers"], errors="coerce")

# For literacy_df, the column is 'literacy_rate'
literacy_df["literacy_rate"] = pd.to_numeric(literacy_df["literacy_rate"], errors="coerce")

# Display the data types after conversion
print("Data types after converting to numeric:")
print("\npopulation_df:")
print(population_df.dtypes)
print("\ngdp_df:")
print(gdp_df.dtypes)
print("\ninternet_users_df_numeric:")
print(internet_users_df_numeric.dtypes)
print("\nliteracy_df:")
print(literacy_df.dtypes)
```

```python
#  Analyze Data in Google Colab

# --- Step 1: Import libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import xml.etree.ElementTree as ET
from functools import reduce # Import reduce here

# --- Step 2: Upload files ---
# from google.colab import files

# print(" Please upload your data files:")
# uploaded = files.upload()

# Example expected files:
# - population.csv
# - gdp.xlsx
# - internet_users.json
# - literacy_rate.xml

# --- Step 3: Load datasets ---
# Population (CSV)
csv_file = "population.csv"
df_population = pd.read_csv(csv_file)   # columns: country, population
# Adjust column names if needed based on actual file header - checking current df_population state
if 'United States' in df_population.columns and '1278642419' in df_population.columns:
  df_population = df_population.rename(columns={'United States': 'country', '1278642419': 'population'})


# GDP (Excel)
excel_file = "gdp.xlsx"
df_gdp = pd.read_excel(excel_file)              # columns: Country, GDP
# Adjust column names if needed based on actual file header - checking current df_gdp state
if 'United States' in df_gdp.columns and '10450436373455' in df_gdp.columns:
   df_gdp = df_gdp.rename(columns={'United States': 'Country', '10450436373455': 'GDP'})


# Internet users (JSON)
json_file = "internet_users.json"
with open(json_file, "r") as f:
    internet_users_data = json.load(f)

# Convert JSON to DataFrame with correct structure (based on successful previous approach)
# Assuming the JSON is a list of dictionaries like [{"country": "USA", "internet_users": 123}, ...]
# If the JSON structure is different (like in the previous notebook state where keys were countries),
# the conversion logic needs to be adjusted. Reverting to list of dicts assumption for now.
# Based on the variable explorer, the JSON data is a list of dicts.

df_internet = pd.DataFrame(internet_users_data)
df_internet = df_internet.rename(columns={'country': 'Country', 'internet_users': 'InternetUsers'})


# Literacy (XML)
xml_file = "literacy_rate.xml"
tree = ET.parse(xml_file)
root = tree.getroot()

literacy_list = []
for country_elem in root.findall("country"): # Renamed variable to avoid conflict
    name = country_elem.get("name") # Get the country name from the 'name' attribute
    literacy_element = country_elem.find("literacy_rate") # Find the literacy_rate element
    literacy = literacy_element.text if literacy_element is not None else None # Get text if element exists
    literacy_list.append({"country": name, "literacy_rate": literacy}) # Use 'country' and 'literacy_rate' as column names
df_literacy = pd.DataFrame(literacy_list)


# --- Step 4: Merge datasets ---
# Standardize country names before merging if necessary (based on previous successful standardization)
def standardize_country_names(df, col):
    mapping = {
        "United States of America": "USA",
        "United Kingdom": "UK",
        "Russian Federation": "Russia",
        "Republic of Korea": "South Korea",
        "Viet Nam": "Vietnam",
        "United States": "USA"
    }
    if col in df.columns:
      df[col] = df[col].replace(mapping)
    return df

df_population = standardize_country_names(df_population, "country")
df_gdp = standardize_country_names(df_gdp, "Country") # gdp_df has 'Country'
df_internet = standardize_country_names(df_internet, "Country") # df_internet will have 'Country'
df_literacy = standardize_country_names(df_literacy, "country")


# Merge using appropriate left_on and right_on for different column names
# Start with population_df
merged_df = df_population.merge(df_gdp, left_on="country", right_on="Country", how="outer")

# Merge with internet data
merged_df = merged_df.merge(df_internet, left_on="country", right_on="Country", how="outer")

# Merge with literacy data
merged_df = merged_df.merge(df_literacy, on="country", how="outer")


# Drop redundant country columns from merges
merged_df = merged_df.drop(columns=['Country_x', 'Country_y'], errors='ignore')


# --- Step 5: Create derived columns ---
# Ensure columns exist and are numeric before calculation
merged_df["InternetUsers"] = pd.to_numeric(merged_df["InternetUsers"], errors="coerce")
merged_df["population"] = pd.to_numeric(merged_df["population"], errors="coerce")
merged_df["GDP"] = pd.to_numeric(merged_df["GDP"], errors="coerce") # Use GDP column from gdp_df
merged_df["literacy_rate"] = pd.to_numeric(merged_df["literacy_rate"], errors="coerce")


merged_df["internet_penetration_pct"] = (merged_df["InternetUsers"] / merged_df["population"] * 100).round(2)
merged_df["gdp_per_capita_usd"] = (merged_df["GDP"] / merged_df["population"]).round(2) # Use GDP column

# --- Step 6: Analysis ---
# 1. Top countries by internet penetration
top_internet = merged_df.sort_values("internet_penetration_pct", ascending=False).head(10)

# 2. Average literacy rate
average_literacy = merged_df["literacy_rate"].mean()

# 3. Correlations
corr_columns = ["gdp_per_capita_usd","population","literacy_rate","internet_penetration_pct"]
corr_matrix = merged_df[corr_columns].corr().round(3)

# --- Step 7: Display results ---
print(" Top 10 Countries by Internet Penetration:\n",
      top_internet[["country","internet_penetration_pct"]].to_string(index=False))
print("\n Average literacy rate across countries: {:.2f}%".format(average_literacy if not pd.isna(average_literacy) else 0)) # Handle potential NaN average
print("\n Correlation matrix:\n", corr_matrix)

# --- Step 8: Visualization ---
# GDP per capita vs Internet penetration
plt.figure(figsize=(6,4))
plt.scatter(merged_df["gdp_per_capita_usd"], merged_df["internet_penetration_pct"])
plt.xlabel("GDP per capita (USD)")
plt.ylabel("Internet penetration (%)")
plt.title("GDP per capita vs Internet penetration")
plt.show()

# Literacy vs Internet penetration
plt.figure(figsize=(6,4))
plt.scatter(merged_df["literacy_rate"], merged_df["internet_penetration_pct"])
plt.xlabel("Literacy rate (%)")
plt.ylabel("Internet penetration (%)")
plt.title("Literacy vs Internet penetration")
plt.show()

# Population vs Internet penetration
plt.figure(figsize=(6,4))
plt.scatter(merged_df["population"], merged_df["internet_penetration_pct"])
plt.xlabel("Population")
plt.ylabel("Internet penetration (%)")
plt.title("Population vs Internet penetration")
plt.show()

# Save merged dataset
merged_df.to_csv("merged_countries_analysis.csv", index=False)
print("\n Merged dataset saved as merged_countries_analysis.csv")
```
